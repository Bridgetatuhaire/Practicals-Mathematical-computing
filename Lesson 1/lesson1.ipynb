Binomial distribution

```python
# Binomial distribution
# Binomial  distribution is a discrete distribution
#it describes the outcome of binary scenarios, toss of a coin
#it has thre parameters
#n-number of trials
#probability occurence of each trials (e.g toss of a coin)
#size - the shape of the returned array
```

```python
#example
#Given 10 trials for coin toss generate 10 data points:
import numpy as np
from numpy import random

x = random.binomial(n=10, p=0.5, size=10)
print(x)
```

Visualiztion of Binomial distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False)
plt.show()
```

Normal distribution

```python
#Example
#Generate a random distribution of size 2x3
import numpy as np
```

```python
from numpy import random
x = random.normal(size=(2,3))
print(x)
```

```python
#Example 2
#Generate a random normal distribution of size 2x3 with mean at 1 and standard deviation of 2
#loc=mean scale = standard deviation
from numpy import random
x = random.normal(loc=1, scale=2, size=(2,3))
print(x)
```

Visualization of normal distribution

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.normal(size=1000), hist=False)
plt.show()
```

```python
sns.distplot(random.normal(size=1000), hist=True)
plt.show()
```

Difference between normal and binormial distribution

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.distplot(random.normal( loc=50, scale=5, size=1000), hist=False, label='normal')
sns.distplot(random.binomial( n=100, p=0.5, size=1000), hist=False, label='binomial')
plt.legend()
plt.show()
```

POISSON DISTRIBUTION

```python
#Poisson Distribution
#it has two parameters:

#Lam - rate or known number of occurrence eg 2 for above problem lam means lambda
#Size - the shape of the returned array
```

```python
#Example: Generate a random 1 x 10 distribution for occurences 2

from numpy import random
x = random.poisson(lam=2, size=10)
print(x)
```

VISUALIZATION OF POISSON DISTRIBUTION

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.poisson(lam=2, size=1000), kde=False)
plt.show()
```
